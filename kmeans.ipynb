```python
import numpy as np
import pandas as pd
import tensorflow as tf
from matplotlib import pyplot as plt

import sys

#tf.enable_eager_execution()
```

```python
# data
data_dir = "../preprocessed_data.csv"
df = pd.read_csv(data_dir)

# splits the data to "training" and "test" data
training_data_pct = 0.8
training_data_n = int(training_data_pct * df.shape[0])
test_data_n = df.shape[0] - training_data_n

df_training_full = df.head(training_data_n)
df_test_full = df.tail(test_data_n)

drop_cols = ["ticker_name", "Date", "Volume"]
label_cols = ['next_day_open', df.columns[-2]]
df_training_clipped = df_training_full.drop(columns=drop_cols)
df_test_clipped = df_test_full.drop(columns=drop_cols)

df_training_arr = df_training_clipped.values
df_test_arr = df_test_clipped.values


#df_arr = df.values
print(df_training_arr.shape)
print(df_test_arr.shape)
print(df_training_full.columns)
df_training_full.describe()
```

```python
num_clusters = 24 # change to 20
evaluation_use_test_data = True
evaluate_labels_data = True

feature_columns = []
for col in df_training_clipped.columns:
    if col not in label_cols:
        feature_columns.append(tf.feature_column.numeric_column(col))
#print(feature_columns)

# TODO: to use mini_batch
kmeans = tf.contrib.factorization.KMeansClustering(
	num_clusters=num_clusters,
    feature_columns=feature_columns,
    use_mini_batch=False)
    #use_mini_batch=True,
    #mini_batch_steps_per_iteration=1000)
```

```python
# train
num_iterations = 10
previous_centers = None
for _ in range(num_iterations):
	kmeans.train(input_training_fn)
	cluster_centers = kmeans.cluster_centers()
	if previous_centers is not None:
		print('delta:', cluster_centers - previous_centers)
	previous_centers = cluster_centers
	print('score:', kmeans.score(input_training_fn))
print('cluster_centers:', cluster_centers)
print('with shape:', cluster_centers.shape)
```

```python
# make labels data if not evaluated in kmeans
def labels_array(use_test_data=True):
    labels_arr = None
    if use_test_data:
        labels_arr = df_test_full[label_cols].values
    else:
        labels_arr = df_training_full[label_cols].values
    
    return labels_arr
        
labels_array(use_test_data=True)
```

```python
# evaluate error
cluster_RMSE = 0
def error_evaluation(use_test_data=True, labels_data_evaluated=False): # if False, use training data
    cluster_indices = list(kmeans.predict_cluster_index(input_test_fn if use_test_data else input_training_fn))
    #pd.DataFrame(cluster_indices)
    
    cluster_density = pd.DataFrame(cluster_indices)[0].value_counts().sort_index()
    print("cluster_density:\n", cluster_density)
    
    labels_arr = None
    if labels_data_evaluated:
        labels_arr = df_test_arr[:,-2:] if use_test_data else df_training_arr[:, -2:]
    else:
        labels_arr = labels_array(use_test_data)
        # create cluster centers labels by averaging
        cluster_centers_data = np.array([]).reshape(num_clusters, 2, 0)
        for i, i_clus in np.ndenumerate(cluster_indices):
            np.insert(cluster_centers_data)
            #cluster_centers_data[i_clus] += labels_arr[i]
        cluster_centers_labels = (cluster_centers_labels.T / cluster_density.values.T).T
        print(cluster_centers_labels)
        
    num_data = labels_arr.shape[0]
    error = []
    
    sum_squared_error = 0
    cluster_sum_squared_error = list(range(num_clusters))
    
    if labels_data_evaluated:
        for i, i_clus in np.ndenumerate(cluster_indices):
            # count RMSE
            error.append(labels_arr[i] - cluster_centers[i_clus][-2:])
            sum_squared_error += error[i[0]] ** 2
            cluster_sum_squared_error[i_clus] += error[i[0]] ** 2
            
        RMSE = (sum_squared_error / num_data) ** 0.5
        print("\nRMSE: ", RMSE)

        cluster_sum_squared_error = np.array(cluster_sum_squared_error)
        global cluster_RMSE
        cluster_RMSE = (cluster_sum_squared_error / num_data) ** 0.5
        cluster_RMSE = pd.DataFrame(cluster_RMSE)
        print("\ncluster_RMSE:\n", cluster_RMSE)

        error = pd.DataFrame(error)
        print("\nerror distribution:\n", error.describe())
    else:
        pass
        
    
    
    
    # evaluate standard deviation misses
    
    error_mean = error.mean()[0]
    error_std = error.std()[0]
    z_threshold = 1 # 95% confidence interval
    error_over_std_dev = abs((error - error_mean) / error_std) > z_threshold
    pct_over_std_dev = error_over_std_dev[0].value_counts()[True] / num_data * 100
    print("%.4fpct of the data are outside of Z threshold (Z < (-%.2f) & Z > (%.2f))" % (pct_over_std_dev, z_threshold, z_threshold))
    
    
    error.hist(bins=100)

error_evaluation(evaluation_use_test_data, evaluate_labels_data)
print()
```

```python
# prepare tp_day and open_day data if necessary
def exp_profit_pct(tp_day, open_day):
    exp_profit_pct = (tp_day - open_day) * 100
    exp_profit_pct = exp_profit_pct.reshape(len(cluster_centers))
    print(exp_profit_pct.shape)
    return exp_profit_pct

open_day = None
tp_day = None
if evaluate_labels_data:
    open_day = np.array([cluster_centers[:,-2]]).T
    tp_day = np.array([cluster_centers[:,-1]]).T
else:
    open_day = "capeeeeee"

exp_profit_pct = exp_profit_pct(tp_day, open_day)
```

```python
# prepare cluster centers data for candlestick chart
def cluster_centers_arr(only_features_price=False):
    if not only_features_price:
        cc = cluster_centers[:,:-2] # drop open day & tp_day

        for _ in range(4):
            cc = np.append(cc, open_day, axis=1)
        for _ in range(4):
            cc = np.append(cc, tp_day, axis=1)
        print(cc.shape)
    else:
        cc = cluster_centers
    
    cc = cc.reshape(len(cluster_centers), int(len(cc[0]) / 4), 4)
    return cc

cc = cluster_centers_arr(only_features_price=not evaluate_labels_data)

arr = [np.arange(len(cluster_centers)), np.arange(cc[0].shape[0])]
this_index = pd.MultiIndex.from_product(arr)

df_cluster_centers = pd.DataFrame()

for entry in cc:
    df_cluster_centers = df_cluster_centers.append(pd.DataFrame(entry), ignore_index=True, sort=False)
df_cluster_centers.index = this_index
df_cluster_centers.columns = ['O', 'H', 'L', 'C']
df_cluster_centers
```

```python
# make candlestick plot
import mpl_finance as pltf

def candlestick_plot(use_test_data=True):
    cluster_indices = list(kmeans.predict_cluster_index(input_test_fn if use_test_data else input_training_fn))
    cluster_density = pd.DataFrame(cluster_indices)[0].value_counts()

    fig, axes = plt.subplots(nrows=1, ncols=cc.shape[0], sharey=True, figsize=(cc.shape[0]*4, 10))
    
    major_gridlines = np.arange(-1, 1, step=0.05)
    minor_gridlines = np.arange(-1, 1, step=0.01)
    
    for i, entry in df_cluster_centers.groupby(level=0):
        pltf.candlestick2_ohlc(axes[i], entry["O"], entry["H"], entry["L"], entry["C"], width=0.75)
        axes[i].set_title("%i - %i items (%.4f pct)\nRMSE_open: %.5fpct\nRMSE_TP: %.5fpct\n\nExpected Profit: %.2fpct"
                          % (i, 
                             cluster_density[i], 
                             cluster_density[i]/df_test_arr.shape[0] * 100, 
                             cluster_RMSE.loc[i,0] * 100, 
                             cluster_RMSE.loc[i,1] * 100,
                             exp_profit_pct[i]
                            ))
        axes[i].set_yticks(major_gridlines)
        axes[i].set_yticks(minor_gridlines, minor=True)
        
        axes[i].grid(which="major", axis="y", alpha=0.5)
        axes[i].grid(which="minor", axis="y", alpha=0.2)
        
        axes[i].tick_params(axis='y', labelleft=True)
        #print(entry["O"])
        #ax = fig.add_subplot(i+1, 1, i+1)
        pass

    fig.savefig("../figures/stock-kmeans.png", format="png")
candlestick_plot(evaluation_use_test_data)
```
